```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import plotly.graph_objects as go
```

```python
day_df = pd.read_pickle('data/LCL_unstack_day.pkl')
day_df.head()
```

```python
Std_test = pd.DataFrame()
Std_train = pd.DataFrame()

Std_test['ds'] = day_df['Std'].tail(round(0.2*day_df["Std"].count())).index.values
Std_test['y'] = day_df['Std'].tail(round(0.2*day_df["Std"].count())).values


Std_train['ds'] = day_df['Std'].loc[day_df.index < Std_test.ds[0]].index.values
Std_train['y'] = day_df['Std'].loc[day_df.index < Std_test.ds[0]].values
```

```python
Std_train_pred = Std_train[['ds']]
Std_test_pred = Std_test[['ds']]
```

```python
prophet = Prophet(yearly_seasonality=True, weekly_seasonality=True, seasonality_mode='multiplicative')
prophet.fit(Std_train)
```

```python
Std_train_pred = prophet.predict(Std_train_pred)
Std_train_pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
Std_test_pred = prophet.predict(Std_test_pred)
Std_test_pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
